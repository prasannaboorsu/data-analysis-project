## Bivariate Analysis

```python
sns.barplot(df,y='Total Price/night(Rs)',x='City')
```

`Interpretation:`
As we compare the  total prices including taxes and charges of cities,we can notice that the average prices are high in mumbai city and low in chennai city.Hyderabad and bangalore cities have moderate prices compared to other cities


```python
sns.boxplot(y="City", x="Price(Rs)", data=df)
```

`Interpretaion` : In the above box plot we can see that there are outliers in  prices of every city,which means when we calculate the box plot,the box represents the interquartile range (IQR) of the data, with a line at the median representing the average price for a hotel in the city per night. The whiskers extend to the most extreme data points within 1.5 times the IQR. The points beyond the whiskers are outliers.
- here the prices in mumbai city are mostly in the range 3000 to 9000.median of the prices is at 4000.The points beyond the upper limit are ouliers
- here the prices in hyderabad and bangalore cities are mostly in the range 2000 to 6000.median of the prices is at 3000.The points beyond the upper limit are ouliers 
- here the prices in channai city are mostly in the range 3000 to 5000.median of the prices is at 4500.The points beyond the upper limit are ouliers

```python
sns.violinplot(y="City", x="Total Price/night(Rs)", data=df)
```

`Interpretation` : We can see that the density of the distribution of data points is dense in the range 0 to 5000 which means mmost of the hotels are in this price range only in all of the cities. The white dot represents the median price of the hotels in the city.The thin line represents the interquartile range of the datapoints.

```python
sns.barplot(x="City", y="No of Reviews", data=df)
```

`Interpretation` : From the bar graph we can notice that hotels in mumbai got more no of reviews and hotels in hyderabad got less no of reviews.The count specifies how accurate are the ratings of the hotels.It also shows that more people are staying in mumbai hotels compared to other cities.

```python
sns.barplot(x="City", y="Tax+Charges(Rs)", data=df)
```

`Interpretation` : From the graph we can see that the taxes and charges are added more to the price in mumbai hotels and less in hyderabad hotels.The black line represents the highest amount of taxes and charges in that particular city.while searching for hotels with low budget one must avoid these hotels with high taxes and charges.

```python
sns.histplot(x="City", y="Review", data=df, cbar=True)
plt.show()
```

`Interpretation` : In the above plot, the shade of the blue represent the density of the datapoints.The darker color represents the datapoints count is above 70 and lighter represents datapoints are below 20 in each city.we can notice that the chennai city haven't got any reviews as Exceptional which indicates that the hotel service is not so good in hotels of chennai city.Most of the very good reviews are in bangalore city.the count of superb and exceptional is relatively less in all the cities.

```python

df2=df.groupby('City')['Total Price/night(Rs)'].min()

df2.plot.bar()
```

Interpretation : In this graph we plotted the minimum price of the hotels in each city,so one must carry atleast that much amount in order to stay in the respective cities in hotels.
-  In hyderabad minimum of 700 rs must be the budget for a night in hotels
-  In chennai minimum of 500 rs must be the budget for a night in hotels
-  In Bangalore minimum of 500 rs must be the budget for a night in hotels
-  In mumbai minimum of 900 rs must be the budget for a night in hotels

`Interpretation` : In this graph we have plotted the maximum as well as minimum prices of the hotels in each city.The prices of the hotels in each city are in the range of the minimum and maximum values.while opting for a luxury hotel one must opt for hotels with higher price

## Multivariate Analysis

```python
sns.boxplot(df,x='Total Price/night(Rs)',y='City',hue='Review')
```

`Interpretation`  :The box represents the interquartile range (IQR) of the data, with a line at the median. The whiskers extend to the most extreme data points within 1.5 times the IQR. The points beyond the whiskers are outliers.Here even though the review is exceptional the price is not so high but when we the whiskers of fabulous review the price is  very high. the hotels with good review have moderate prices in each city.The value of exceptional review is negligible in mumbai city.The chennai city hotels have no exceptional review at all

```python
sns.stripplot(df,x='City',hue='Tax+Charges(Rs)',y='Price(Rs)',palette='rocket')
```

`Interpretation`: In the above strip,we are comparing the prices and taxes in each city,the hotels with lower prices per night charge less amount of taxes and charges and vice versa.The prices as well as tax charges are relatively high in mumbai city which may be due to the hotels being luxury ones or might be due to better services.

`Interpretation`: In the above plots,we are plotting the count of review in each city using catplot of kind count. Here we can see that the chennai city have no hotel with exceptional review.Also mumbai city has more no of hotels with good review.Both hyderabad and bangalore got equal amount of exceptional review hotels.Hotels in bangalore has more no of hotels with very good review relatively  to the other ities

```python
sns.scatterplot(df,x='Total Price/night(Rs)',y='Price(Rs)',hue='Tax+Charges(Rs)',palette='rocket')
```
